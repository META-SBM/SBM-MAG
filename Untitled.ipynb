# Создадим файл для MUSCLE

Список бинов, прошедших контроль качества

```python
from Bio import SeqIO
bins = filtered_mag
```

```python
bins[1]
```

```python
samples = list(set([elem.split('_')[0] + '_' + elem.split('_')[1].split('.')[0] for elem in bins]))
len(samples)
```

```python
samples[1]
```

```python
fasta_files = []
for sample in samples:
    path = f'/mnt/disk1/PROJECTS/DAVID_WGS/binning/DASTool/dastool_out/{sample}/drep/data/checkM/checkM_outdir/storage/tree/concatenated.fasta'
    fasta_files.append(path)
```

```python
fasta_files[1]
```

```python

# Путь к выходному файлу
output_fasta_path = '/mnt/disk1/PROJECTS/DAVID_WGS/binning/DASTool/dastool_out/genes_concat_muscle.fasta'  # Замените на желаемый путь к выходному файлу

# Открываем выходной файл для записи
with open(output_fasta_path, 'w') as output_handle:
    # Проходим по каждому FASTA файлу
    for fasta_file in fasta_files:
        try:
            # Читаем записи из FASTA файла
            for record in SeqIO.parse(fasta_file, "fasta"):
                # Проверяем, содержится ли ID записи в списке бинов
                if record.id in bins:
                    # Записываем запись в выходной файл
                    SeqIO.write(record, output_handle, "fasta")
        except FileNotFoundError:
            print(f"Файл не найден: {fasta_file}")
        except Exception as e:
            print(f"Произошла ошибка при обработке {fasta_file}: {e}")

print(f"Фильтрация завершена. Результаты сохранены в {output_fasta_path}.")
```
